## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

## Ejercicio 4
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C 
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1

print(A == B)    # 1. Si A y B son equivalentes
print(A != C)    # 2. Si A y C NO son equivalentes
print(A > C)     # 3. Si A es mayor que C
print(C <= A)    # 4. Si C es menor o igual que A
print(B != C)    # 5. Si B NO es equivalente a C
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input1 = input("Introduce el primer valor: ")
input2 = input("Introduce el segundo valor: ")
input3 = input("Introduce el tercer valor: ")

todos_iguales = input1 == input2 == input3
dos_iguales = input1 == input2 or input1 == input3 or input2 == input3

print("Todos son iguales:", todos_iguales)
print("Todos son iguales:", dos_iguales)
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

````python
input1 = float(input("Introduce el primer número: "))
input2 = float(input("Introduce el segundo número: "))

suma = input1 + input2

print("Es mayor que 10:", suma > 10)
print("Es menor que 10:", suma < 10)
print("Es igual que 10:", suma == 10)
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
frase = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

print(frase.upper())
print(frase.lower())
print(frase.title())
print(frase.split())
print(frase.replace(",", ";"))
print(frase.replace("a", ""))
````
