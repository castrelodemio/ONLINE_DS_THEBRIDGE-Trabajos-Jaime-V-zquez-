Crea una matriz aleatoria de 5x5 y estandariza sus valores. Para normalizar tienes que restar sus valores por la media(`np.mean`) y dividir el resultado por la desviación estándar(`np.std`).

## Ejercicio 10
1. Crea una secuencia de números que vayan del 0 al 10.
2. Crea un nuevo array, que parta del anterior, pero sustituya todos los elementos entre el 3 y el 8, por el valor -1
3. Crea un nuevo array que niegue todos los elementos entre el 3 y el 8

````python
arr = np.arange(0, 11)
print(arr)

nuevo = np.where((arr > 3) & (arr < 8), -1, arr)
print(nuevo)

negado = np.where((arr > 3) & (arr < 8), arr, -arr)
print(negado)
````

## Ejercicio 11
#### Bonus track. Mini hundir la flota
1. Crea un tablero de 10x10 relleno del carácter espacio ( ).
2. Posiciona un par de barcos en [(0,1), (1,1)] y [(1,3), (1,4), (1,5), (1,6)]. Los barcos serán Os mayúsculas. Como ves, un barco de dos posiciones de eslora y otro de cuatro.
3. Recibe un disparo en uno de los barcos, sustituyendo la O por una X
4. Recibe un disparo en agua, sustituyendo uno de los espacios por un guión: -
5. Ignorando los disparos recibidos, ahora prueba a posicionar otro barco de 4 posiciones dentro del tablero, pero **de manera ALEATORIA.**

**Consejo apartado 5**: El barco tendrá una posición inicial aleatoria dentro del tablero, y una orientación también aleatoria (N,S,E,O).

Por ejemplo, si tu posicion inicial fuese (5,0) y la orientación O, no podrias ponerlo porque se sale del tablero. Ahora bien, si fuese orientación este, si podrías, y ocuparía las posiciones [(5,0), (5,1), (5,2), (5,3)]. Por tanto tendrás que iterar con un bucle si es posible encajar el barco en el tablero, y el bucle terminará cuando lo consigas encajar. **Ojo con las posiciones donde ya hay barco!!!**

Usa `import random` para elegir las coordenadas aleatorias.

````python
tablero = np.full((10, 10), " ")

tablero[0, 1] = "O"
tablero[1, 1] = "O"
tablero[1, 3] = "O"
tablero[1, 4] = "O"
tablero[1, 5] = "O"
tablero[1, 6] = "O"

tablero[0, 1] = "X"
tablero[0, 0] = "-"

tablero
````

````python
import random

tablero2 = np.full((10, 10), " ")

eslora = 4
colocado = False

while not colocado:
    fila = random.randint(0, 9)
    col = random.randint(0, 9)
    orientacion = random.choice(["N", "S", "E", "O"])

    if orientacion == "E" and col + eslora <= 10:
        posiciones = [(fila, col + i) for i in range(eslora)]
    elif orientacion == "O" and col - eslora >= -1:
        posiciones = [(fila, col - i) for i in range(eslora)]
    elif orientacion == "S" and fila + eslora <= 10:
        posiciones = [(fila + i, col) for i in range(eslora)]
    elif orientacion == "N" and fila - eslora >= -1:
        posiciones = [(fila - i, col) for i in range(eslora)]
    else:
        continue

    if all(tablero2[f, c] == " " for f, c in posiciones):
        for f, c in posiciones:
            tablero2[f, c] = "O"
        colocado = True

tablero2
````
