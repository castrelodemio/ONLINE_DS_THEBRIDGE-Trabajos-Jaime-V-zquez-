# Occupation

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv(url, sep="|", index_col="user_id")
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")["age"].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.groupby("occupation")["gender"].count()
```

```python
users.groupby("gender").count()
```

```python
users["gender_n"] = (users["gender"] == "M").astype(int)
users
```

```python
def ratio(x):
    print("tipo de x:", type(x))
    x["ratio"] = x["gender_n"].sum() / len(x)
    return x

users = users.groupby("occupation").apply(ratio)
users
```

```python
users = users.groupby("occupation").apply(lambda x: x.assign(ratio=x["gender_n"].sum() / len(x)))
users
```

```python
users.groupby("occupation")["gender_n"].sum()
```

```python
users.groupby("occupation")["age"].count()
```

```python
male_ratio = users.groupby("occupation")["gender"].apply(lambda x: (x == "M").mean() * 100).sort_values(ascending=False)
male_ratio.map(lambda x: f"{round(x, 2)}%")
```

```python
users.groupby("occupation")["gender"].apply(lambda x: (x == "M").mean()).sort_values(ascending=False).map(lambda x: f"{round(x*100, 2)}%")
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].agg(["min", "max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"])[["age"]].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
users.groupby(["occupation", "gender"]).size().reset_index(name="count")
```

```python
users.groupby("occupation")["gender"].value_counts(normalize=True).mul(100).sort_values(ascending=False)
```
