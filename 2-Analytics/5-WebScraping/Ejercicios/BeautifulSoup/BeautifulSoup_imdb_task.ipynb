```python
import pandas as pd
import json
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
import time

url = "https://www.imdb.com/chart/top/"

driver = uc.Chrome(
    browser_executable_path=r"C:\Program Files (x86)\Google\Chrome\Application\chrome.exe",
    headless=False,
    version_main=147
)

driver.get(url)
time.sleep(5)

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
print(soup.title)
```

```python
print(html[:2000])
```

```python
n = 0
for x in soup.find_all("h3", class_="ipc-title__text"):
    print(x)
    n += 1
    print(n)
```

```python
for x in soup.find_all("h3", class_="ipc-title__text")[0:25]:
    #print("Ranking:", x.get_text().split(".")[0])
    print("Título:", x.get_text())
    #print(x)
```

```python
for x in soup.find_all("div", class_="ipc-signpost__text")[0:26]:
    #print("Ranking:", x.get_text().split(".")[0])
    print("Ranking:", int(x.get_text().replace("#","")))
    #print(x)

```

```python
for x in soup.find_all('span', class_="ipc-rating-star--rating"):
    print("Rating:", x.get_text())
```

```python
for x in soup.find_all("span", class_="sc-3ac6b047-6 dTZPpv cli-title-metadata-item"):
    #print(x)
    text = x.get_text()
    if len(text) == 4:
        print("Año:", text)
    if "h" in text:
        print("Duración:", text)
```

```python
print("Ranking:", len([int(x.get_text().replace("#","")) for x in soup.find_all("div", class_="ipc-signpost__text")][0:25]))
print("Título:", len([x.get_text() for x in soup.find_all("h3", class_="ipc-title__text")][0:25]))
print("Año:", len([x.get_text() for x in soup.find_all("span", class_="sc-3ac6b047-6 dTZPpv cli-title-metadata-item") if len(x.get_text()) == 4]))
print("Duración:", len([x.get_text() for x in soup.find_all("span", class_="sc-3ac6b047-6 dTZPpv cli-title-metadata-item") if "h" in x.get_text()]))
print("Rating:", len([x.get_text() for x in soup.find_all('span', class_="ipc-rating-star--rating")]))
```

```python
dict_peliculas = {"Ranking": [int(x.get_text().replace("#","")) for x in soup.find_all("div", class_="ipc-signpost__text")][0:25],
                  "Título": [x.get_text() for x in soup.find_all("h3", class_="ipc-title__text")][0:25],
                  "Año": [x.get_text() for x in soup.find_all("span", class_="sc-3ac6b047-6 dTZPpv cli-title-metadata-item") if len(x.get_text()) == 4],
                  "Duración": [x.get_text() for x in soup.find_all("span", class_="sc-3ac6b047-6 dTZPpv cli-title-metadata-item") if "h" in x.get_text()],
                  "Rating": [x.get_text() for x in soup.find_all('span', class_="ipc-rating-star--rating")]
}

df = pd.DataFrame(dict_peliculas)
df.to_csv("top25.csv")
df
```
